```python
import random
import json
import xml.etree.ElementTree as ET


def generate_weather_data():
    cities = ["London", "Manchester", "Ottawa", "New York", "Tokyo"]
    conditions = ["Sunny", "Cloudy", "Rainy", "Snowy"]

    weather_data = []

    for city in cities:
        weather_data.append({
            "city": city,
            "temperature": round(random.uniform(-5, 35), 1),
            "humidity": random.randint(20, 90),
            "condition": random.choice(conditions)
        })

    return weather_data


def get_weather_json(weather_data):
    return json.dumps({"weather": weather_data}, indent=4)


def get_weather_xml(weather_data):
    root = ET.Element("WeatherData")

    for entry in weather_data:
        city_elem = ET.SubElement(root, "City")
        ET.SubElement(city_elem, "Name").text = entry["city"]
        ET.SubElement(city_elem, "Temperature").text = str(entry["temperature"])
        ET.SubElement(city_elem, "Humidity").text = str(entry["humidity"])
        ET.SubElement(city_elem, "Condition").text = entry["condition"]

    return ET.tostring(root, encoding="unicode")


if __name__ == "__main__":
    data = generate_weather_data()

    print("JSON Output:")
    print(get_weather_json(data))

    print("\nXML Output:")
    print(get_weather_xml(data))
```
